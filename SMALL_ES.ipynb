```python
def get_metrics(mysearch,elsearch):
    precision_list=[]
    recall_list=[]
    f1_list=[]
    acc_list=[]
    qresponse_list1=[]
    qresponse_list2=[]
    for q in mysearch:
        Retrieved=set(mysearch[q]['hits'])
        Relevant=set(elsearch[q]['hits'])
        NonRelevant=(set(data.keys())).difference(Relevant)
        TP=Retrieved.intersection(Relevant)
        FP=Retrieved.difference(TP)
        FN=Relevant.difference(TP)
        TN=NonRelevant.difference(FP)
        
        print("######",len(TP),len(FP),len(TN),len(FN))
        
        precision=len(TP)/(len(TP)+len(FP))
        recall=len(TP)/(len(TP)+len(FN))
        f1=2*precision*recall/(precision+recall)
        acc=(len(TP)+len(TN))/(len(TP)+len(FP)+len(TN)+len(FN))
        
        #precision=len(TP)/len(Relevant)
        
        qresponse_list1.append(elsearch[q]['took'])
        qresponse_list2.append(mysearch[q]['took'])
        precision_list.append(precision)
        recall_list.append(recall)
        f1_list.append(f1)
        acc_list.append(acc)
    print("*********\nUnranked Evaluation")
    print("Avg precision is:\t",sum(precision_list)/len(precision_list))
    print("Avg recall is:\t",sum(recall_list)/len(recall_list))
    print("Avg f1 is:\t",sum(f1_list)/len(f1_list))
    print("Avg Acc is:\t",sum(acc_list)/len(acc_list))
    print("Avg Query Response in (ms) ELASTIC is:\t",sum(qresponse_list1)/len(qresponse_list1))
    print("Avg Query Response in (ms) MYSEARCH is:\t",sum(qresponse_list2)/len(qresponse_list2))
    
    
    
import json
mysearch_output=json.load(open("SmallOutput/mysearch_output1.json"))
elsearch_output=json.load(open("SmallOutput/elsearch_output1.json"))
get_metrics(mysearch_output,elsearch_output)
```
